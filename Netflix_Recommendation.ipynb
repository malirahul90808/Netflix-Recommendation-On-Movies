```python
def fetch_cast_name(data):

    counter = 0
    lst = []
    for i in ast.literal_eval(data):
        if counter != 3 :
           lst.append(i['name'])
           counter += 1
        else:
            break

    return lst
```

NLP Concept

```python
# tokenization
# lower case
# stemming
# stopwords removal
# encoding into vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ps = PorterStemmer()

def stem(text):

    stemmed = []

    for i in text.split():
        stemmed.append(ps.stem(i))

    return " ".join(stemmed)
```

```python
stem(new_df.tags[0])
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

Final Function to Recommend Top 5 similar Movies

```python
def recommend(movie):

    movie_index = new_df[new_df['title'] == movie].index[0]  # Finding the index of movie for that we want to get recommendation
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    print('Top 5 recommended movies are:')
    print(' ')
    for movie in movie_list:

        print(new_df.iloc[movie[0]].title)  # Printing the movie title form the index of similar movies
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```

```python
recommend('The R.M.')
```
